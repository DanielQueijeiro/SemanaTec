```python
# Carga las librerías necesarias.
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

El conjunto de datos contiene información demográfica sobre los asegurados en una compañía de seguros:
* **Age**: Edad de la persona.
* **Average of ap_hi**: Promedio de presión alta.
* **Average of ap_lo**: Promedio de presión baja.

```python
# Grafica la información de la edad y presión alta
fig = plt.figure(figsize=(10,10))
sns.stripplot( data = df, y ='Average of ap_hi', x = 'Age')
```

Genera una regresión líneal para obtener una aproximación de la ecuación 
$$ y = ax + b $$
donde $ a $ se conoce comúnmente como **pendiente**, y $ b $ se conoce comúnmente como **intersección**, tanto para presión alta como la presión baja.

```python
# ¿Cuál es el valor de a y cuál es el valor de b para la presión alta?
x_hi = df['Average of ap_hi']
x_hi = np.array(x_hi)
y_hi = df['Age']
y_hi = np.array(y_hi)


model = LinearRegression(fit_intercept=True)
model.fit(x_hi[:, np.newaxis], y_hi)

xfit_hi = np.linspace(110,145,50)
yfit_hi = model.predict(xfit_hi[:, np.newaxis])

plt.scatter(x_hi,y_hi)
plt.plot(xfit_hi, yfit_hi, color='green');
plt.plot(x_hi,y_hi, 'o')
plt.plot(np.vstack([x_hi,x_hi]), np.vstack([y_hi, model.predict(x_hi[:, np.newaxis])]), color = 'red');
plt.title('Regresión lineal')
plt.xlabel('Presión')
plt.ylabel('Edades')


print('Pendiente del modelo (a): ', model.coef_[0])
print('Intercepción del modelo (b): ', model.intercept_)
```

```python
# ¿Cuál es el valor de a y cuál es el valor de b para la presión baja?
x_lo = df['Average of ap_lo']
x_lo = np.array(x_lo)
y_lo = df['Age']
y_lo = np.array(y_lo)


model = LinearRegression(fit_intercept=True)
model.fit(x_lo[:, np.newaxis], y_lo)

xfit_lo = np.linspace(70,110,50)
yfit_lo = model.predict(xfit_lo[:, np.newaxis])

plt.scatter(x_lo,y_lo)
plt.plot(xfit_lo, yfit_lo, color='green');
plt.plot(x_lo,y_lo, 'o')
plt.plot(np.vstack([x_lo,x_lo]), np.vstack([y_lo, model.predict(x_lo[:, np.newaxis])]), color = 'red');
plt.title('Regresión lineal')
plt.xlabel('Presión')
plt.ylabel('Edades')

print('Pendiente del modelo (a): ', model.coef_[0])
print('Intercepción del modelo (b): ', model.intercept_)
```
